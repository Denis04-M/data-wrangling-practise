**Merging DataFrames**

```python
df1 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                      'data1': range(7)})
df1
```

```python
df2 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
df2
```

```python
pd.merge(df1, df2, left_on='lkey', right_on='rkey')
```

```python
df_left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
              'key2': ['one', 'two', 'one'],
              'lval': [1, 2, 3]})
df_left
```

```python
df_right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                    'key2': ['one', 'one', 'one', 'two'],
                    'rval': [4, 5, 6, 7]})
df_right
```

```python
pd.merge(df_left, df_right, on=['key1', 'key2'], how='outer')
```

```python
left_df = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],'value': range(6)})
left_df
```

```python
right_df = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
right_df
```

```python
# merging using index
pd.merge(left_df, right_df, left_on='key', right_index=True, how='outer')
```

```python
import pandas as pd
import pandas_datareader as web

start_date = '2007-01-01'
end_date = '2012-01-01'

data1 = web.DataReader(['SPX', 'AMZN', 'FORD'], 'yahoo', start_date, end_date)
data = data1['Adj Close']
data
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.rc('figure', figsize = (15, 5))

# ax.plot(data['SPX'], 'k--')
```

```python
from datetime import datetime

spx = data['SPX']
spx.plot(ax=ax, style='k-')

crisis_data = [
 (datetime(2007, 10, 11), 'Peak of bull market'),
 (datetime(2008, 3, 12), 'Bear Stearns Fails'),
 (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:
  ax.annotate(label, xy=(date, spx.asof(date) + 2),
  xytext=(date, spx.asof(date) + 2),
  arrowprops=dict(facecolor='black', headwidth=4, width=2,
  headlength=4),
  horizontalalignment='left', verticalalignment='top')

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([0, 5])
ax.set_title('Important dates in the 2008-2009 financial crisis')

fig
```
