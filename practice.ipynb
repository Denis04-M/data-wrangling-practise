```python
import pandas as pd

# how to make a DataFrame

data = {'Migrants': [954806, 543822, 283922, 260650, 242032, 204796, 192456],
        'Population Density': [36, 240, 110, 281, 4, 46, 9],
        'Country': ['USA', 'Germany', 'Turkey', 'UK', 'Canada', 'Columbia', 'Russia'],
        'Population': [331002651, 83783942, 84339067, 67886011, 37742154, 50882889, 142934462]}
pop1 = pd.DataFrame(data)
print(pop1, '\n-------------------------------------------------------')

# rearange colums
pop = pd.DataFrame(pop1, columns=['Country', 'Population', 'Migrants',
                                 'Population Density'])

print(pop, '\n-------------------------------------------------------')

# init. index
print(pd.DataFrame(data, columns=['Country', 'Population', 'Migrants','Population Density'],
                    index=[1, 2, 3, 4, 5, 6, 7]), '\n---------------------------------------')

print(pop.columns)

print(pop.index)
```

**Merging DataFrames**

```python
df1 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                      'data1': range(7)})
df1
```

```python
df2 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
df2
```

```python
pd.merge(df1, df2, left_on='lkey', right_on='rkey')
```

```python
df_left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
              'key2': ['one', 'two', 'one'],
              'lval': [1, 2, 3]})
df_left
```

```python
df_right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                    'key2': ['one', 'one', 'one', 'two'],
                    'rval': [4, 5, 6, 7]})
df_right
```

```python
pd.merge(df_left, df_right, on=['key1', 'key2'], how='outer')
```

```python
left_df = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],'value': range(6)})
left_df
```

```python
right_df = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
right_df
```

```python
# merging using index
pd.merge(left_df, right_df, left_on='key', right_index=True, how='outer')
```
